# Productor para divisas de criptomonedas

Para conocer las diferentes divisas para la criptomoneda de Ethereum, utilizaresmo una api Rest de https://min-api.cryptocompare.com/ donde extraeremos la información cada 5 segundos a traves del modulo `request`. De esta manera creamos un flujo de datos para posterior analisis.

```
producer = KafkaProducer(bootstrap_servers='localhost:9092')

# Obtención de las diferentes divisas
while True:       
    response = requests.get(api_url)
    json_ = response.json()
    print(str(json_["ETH"]).encode('utf-8'))    
    producer.send("criptos", str(json_["ETH"]).encode('utf-8'))   
    time.sleep(5)
```
